# Data visualization with Python- Assigment 

```python
#Assignment 1

import numpy as np 
import pandas as pd 
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = x*x 
x
```

```python
y
```

```python
plt.plot(x,y, 'bo--')
plt.title('Assigment 1', fontsize = 20, color='blue')
plt.xlabel('x-axis', fontsize= 10, color = 'blue')
plt.ylabel('y-axis', fontsize = 10, color ='blue')
```

```python
#Assigment-2

x = [1,2,3,4,5,6,7] #days of the week
y = [160,150,140,145,175,165,180] #sales
x
```

```python
y
```

```python
plt.plot(x,y, 'go-', linewidth= '2')
plt.title('Sales in the week', fontsize='20', color='green')
plt.xlabel('Days of the week', fontsize='10', color='green')
plt.ylabel('Sales', fontsize='10', color='green')
plt.show()
```
